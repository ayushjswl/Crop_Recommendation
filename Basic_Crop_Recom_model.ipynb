```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

soil_enc = LabelEncoder()
season_enc = LabelEncoder()
crop_enc = LabelEncoder()

df['SOIL_ENC'] = soil_enc.fit_transform(df['SOIL'])
df['SEASON_ENC'] = season_enc.fit_transform(df['SEASON'])
df['CROPS_ENC'] = crop_enc.fit_transform(df['CROPS'])

df
```

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

df
```

```python
features = ['SOIL_ENC', 'SEASON_ENC', 'SOIL_PH', 'TEMP', 'RELATIVE_HUMIDITY', 'N', 'P', 'K']

X = df[features]
y = df['CROPS_ENC']
```

```python
import seaborn as sns
sns.heatmap(X.corr(), annot=True, cmap='coolwarm', center=0, linewidths=0.5, linecolor='white')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(crop_enc.classes_)
```

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
rfclassifier = RandomForestClassifier(random_state=42)
rfclassifier.fit(X_train, y_train)
recc_crop_list = rfclassifier.predict_proba(X_test)

top_crops_index = np.argsort(recc_crop_list, axis=1)[:, -5:][:, ::-1]
#print(top_crops_index)
last_crops_index = np.argsort(recc_crop_list, axis=1)[:, :5]
#print(last_crops_index)

top_crops_name = crop_enc.classes_[top_crops_index]
last_crops_name = crop_enc.classes_[last_crops_index]
print(top_crops_name)
print(last_crops_name)
```

```python
import numpy as np
import pandas as pd

unique_crops = list(df['CROPS'].unique())

top_position_counts = {pos: dict.fromkeys(unique_crops, 0) for pos in range(1, 6)}
last_position_counts = {pos: dict.fromkeys(unique_crops, 0) for pos in range(1, 6)}

for row in top_crops_name:
    for pos, crop in enumerate(row, 1):
        top_position_counts[pos][crop] += 1

for row in last_crops_name:
    for pos, crop in enumerate(row, 1):
        last_position_counts[pos][crop] += 1


top_df = pd.DataFrame(top_position_counts).T  # Positions as rows
last_df = pd.DataFrame(last_position_counts).T

print("Top positions count:")
print(top_df)

print("\nLast positions count:")
print(last_df)

top_df.to_csv('top_crop_position_counts.csv')
last_df.to_csv('last_crop_position_counts.csv')
```
